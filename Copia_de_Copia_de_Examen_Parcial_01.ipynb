**Pregunta 1**

Los sueldos en dólares de los empleados de dos empresas, son los siguientes:

**Empresa A:**

|$Sueldos$|$f_i$|
|:--|:--|
|$[70,100>$|$5$|
|$[100,130>$|$8$|
|$[130;160>$|$3$|
|$[160;190>$|$4$|
|$[190;220]$|$1$|

**Empresa B:**

45, 65,  215,  200,  220,  150,  195,  158, 210,  220,  340,  360,  120,  210,  220

a.	Calcular e interpretar la media y desviación estándar de cada empresa.

b.	Calcular el coeficiente de variación de cada empresa.

c.	¿En cuál de las dos empresas sus sueldos son más homogéneas?





**Solución:**

**a. Calcular e interpretar la media y desviación estándar de cada empresa**

**b. Calcular el coeficiente de variación de cada empresa.**


> Empresa A



```
x = c(85,115,145,175,205)
fi = c(5,8,3,4,1)
n = sum(fi)
media = sum(x*fi)/n
varianza = sum((x-media)^2*fi)/(n-1)
desviacion = sqrt(varianza)
variacion = (desviacion/media)*100
titulos = c("Media", "DE","CV")
datos = c(media,desviacion,variacion)
T = data.frame(titulos,datos)
T
```

El sueldo medio es de 127.86 dólares y la desviación estandar se puede inferir que la distribución es asimétrica.



> Empresa B




```
x = c(45, 65, 215, 200, 220, 150, 195, 158, 210, 220, 340, 360, 120, 210, 220)
n = length(x)
media = sum(x)/n
varianza = sum((x-media)^2)/(n-1)
desviacion = sqrt(varianza)
variacion = (desviacion/media)*100
titulos = c("Media", "DE","CV")
datos = c(media,desviacion,variacion)
T = data.frame(titulos,datos)
T
```

El sueldo medio es de 195 dólares y la desviación estandar se puede inferir que la distribución es asimétrica, aún más que la empresa A.

**c ¿En cuál de las dos empresas sus sueldos son más homogéneas?**

Dados los coeficientes de variación podemos inferir en el caso de la empresa A que los sueldos son ligeramente homegeneos, pero en el caso de la empresa B los valores son heterogéneos. Por lo que la empresa A tiene sueldos más homogéneos.

Los axiomas son el conjunto de datos adquiridos.

**Pregunta 3**

Para determinar y analizar una distribución de frecuencias por intervalos, ingresar datos desde una hoja de cálculo,



Use la base de datos correspondiente a la hoja de cálculo denominada **CONSUMO_AGUA** para elaborar su distribución de frecuencias, luego analice sus resultados

**Solución:**

```
x=M$CONSUMO
N=nclass.Sturges(x)
Intervalo=cut(x,breaks = N,right = F)
DF=data.frame(table(Intervalo))
print(DF)
fi=table(Intervalo)
barplot(fi,col="yellow3",space=0,main="Histograma",xlab="valores",horiz="false")
```

Se puede apreciar un sesgo hacia la derecha y una distribución asimétrica de los datos. Se puede inferir que la mayor parte del consumo (más del 70%) es menor que la media del total y son pocos los casos atípicos.
